# 데이터 출처
* Pima Indians 당뇨병 데이터셋(Pima Indians Diabetes Database)
* 캐글 URL : https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database

## 데이터 구성
* Pregnancies : 임신 횟수
* Glucose : 2시간 동안의 경구 포도당 내성 검사에서 혈장 포도당 농도
* BloodPressure : 이완기 혈압 (mm Hg)
* SkinThickness : 삼두근 피부 주름 두께 (mm), 체지방을 추정하는데 사용되는 값
* Insulin : 2시간 혈청 인슐린 (mu U / ml)
* BMI : 체질량 지수 (체중kg / 키(m)^2)
* DiabetesPedigreeFunction : 당뇨병 혈통 기능
* Age : 나이
* Outcome : 768개 중에 268개의 결과 클래스 변수(0 또는 1)는 1이고 나머지는 0입니다.

# 필요한 라이브러리 로드

```python
%config Completer.use_jedi = False
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

# Feature Engineering

## 수치형 변수를 범주형 변수로 만들기

```python
df["Pregnancies_high"] = df["Pregnancies"] > 6
df[["Pregnancies", "Pregnancies_high"]].head()
```

```python
# One-Hot-Encoding
# 수치 => 범주 => 수치
df["Age_low"] = df["Age"] < 30
df["Age_middle"] = (df["Age"] >= 30) & (df["Age"] <= 60)
df["Age_high"] = df["Age"] > 60
df[["Age", "Age_low", "Age_middle", "Age_high"]].head()
```

```python
sns.countplot(data=df, x="Age_low", hue="Outcome")
# 나이가 25세 미만일 때 현저히 발병율이 낮음을 확인
```

```python
sns.countplot(data=df, x="Age_middle", hue="Outcome")
# 나이가 25세 이상, 60세 이하일때는 발병율에 큰 차이가 없음을 확인
```

```python
sns.countplot(data=df, x="Age_high", hue="Outcome")
# 나이가 60세 이상일때는 발병율에 큰 차이가 없음을 확인
```

## 결측치 다루기

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# 0값을 결측치로 만들기
df["Insulin_nan"] = df["Insulin"].replace(0, np.nan)
df[["Insulin", "Insulin_nan"]].head()
```

```python
# 결측치 개수
df["Insulin_nan"].isnull().sum()
```

```python
# 결측치 비율
df["Insulin_nan"].isnull().mean()
```

```python
# 0값을 그대로 둔 원래 컬럼과 결측치 제거한 컬럼의 Outcome 평균 비교
df.groupby(["Outcome"])["Insulin", "Insulin_nan"].mean()
```

```python
# 결측치 대체값으로 처리하기
# 어떤 값으로 대체?
df.groupby(["Outcome"])["Insulin", "Insulin_nan"].agg(["mean", "median"])

# 두 값이 많이 차이남
```

```python
# 결측치 채우기
df.loc[(df["Outcome"] == 1) & (df["Insulin_nan"].isnull()), "Insulin_nan"] = 169.5 # mean으로 채우기
df.loc[(df["Outcome"] == 0) & (df["Insulin_nan"].isnull()), "Insulin_nan"] = 102.5 # mean으로 채우기
```

# 학습, 예측 데이터셋 나누기

```python
# 8:2의 비율로 구하기 위해 전체 데이터의 행에서 80% 위치에 해당하는 값을 구해서 split_count라는 변수에 담는다.
# split_count
split_count = int(df.shape[0]*0.8) # 행의 개수에 0.8 곱하여 80% 위치에 해당하는 행 인덱스 찾기 -> 소수점 제외한 인덱스 값만(int)
split_count
```

```python
# train, test로 슬라이싱을 통해 데이터를 나눈다.
train = df[:split_count].copy()
train
```

```python
# train dateset에서만 이상치 제거
# 600 미만인 데이터만 다시 train에 저장
train = train[train["Insulin_nan"] < 600]
train.shape
```

```python
test = df[split_count:].copy()
test
```

# 학습, 예측에 사용할 컬럼

```python
# feature_names 라는 변수에 학습과 예측에 사용할 컬럼명을 가져온다.
# outcome만 제외하고 가져오기
feature_names = train.columns.tolist()
feature_names.remove("Pregnancies")
feature_names.remove("Outcome")
feature_names.remove("Age_low")
feature_names.remove("Age_middle")
feature_names.remove("Age_high")
feature_names.remove("Insulin")
feature_names.remove("Insulin_log")
feature_names
```

# 정답값이자 예측해야될 값

```python
# label_name 이라는 변수에 예측할 컬럼의 이름을 담는다.
# label_name은 하나이기 때문에 리스트로 만들어주지 않아도 된다.
label_name = "Outcome"
label_name
```

# 학습, 예측 데이터셋 만들기

```python
# 학습 세트 만들기 예) 시험의 기출문제
X_train = train[feature_names]
print(X_train.shape)
X_train.head() # 데이터프레임 형태
```

```python
# 정답 값 만들기 예) 기출문제의 정답
y_train = train[label_name]
print(y_train.shape) # 학습데이터(X_train과 행의 개수 같아야 함.)
y_train.head() # 벡터 형태
```

```python
# 예측에 사용할 데이터세트 만들기 예) 실전 시험 문제
X_test = test[feature_names]
print(X_test.shape)
X_test
```

```python
# 예측의 정답값 예) 실전 시험 문제의 정답
y_test = test[label_name]
print(y_test.shape)
y_test.head()
```

# 머신러닝 알고리즘 가져오기

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42)
model
```

# 학습(훈련)
* 시험을 볼 때 기출문제(X_train)와 정답(y_train)을 보고 공부하는 과정과 유사하다.

```python
model.fit(X_train, y_train)
```

# 예측
* 실전 시험문제(X_test)라고 보면 됨. 우리가 정답을 직접 예측함.

```python
y_predict = model.predict(X_test)
y_predict[:5]
```

# 트리 알고리즘 분석하기
* 의사결정나무를 시각화

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,20))
tree = plot_tree(model,
                 feature_names=feature_names,
                filled=True,
                fontsize=10)
```

```python
# 피처의 중요도 추출하기

model.feature_importances_
```

```python
# 피처의 중요도 시각화

sns.barplot(x=model.feature_importances_, y=feature_names)
```

# 정확도(Accuracy) 측정하기

```python
# 살제값 - 예측값을 빼주면 같은 값은 0으로 나오게 된다.
# 여기에서 절대값을 씌운 값이 1인 값이 다르게 예측한 값이 된다.
# 44 -> 39 -> 51(나이 25세 기준) -> 51(나이 30세 기준) -> 36(인슐린 결측치 평균으로 대체) => 16(인슐린 결측치 중앙값으로 대체)
# => 16(인슐린 로그변환) => 15(파생변수 low_glu_insulin 추가) => 15(인슐린 600이상 이상치 제거) => 15(sclaing)
# 원핫인코딩한다고해서 무조건 모델 성능이 좋아지는 것은 아님

diff_count = abs(y_test - y_predict).sum()
print(diff_count) # 39개가 잘못 예측됨.
diff_count / len(y_test) # 25를 잘못 예측함. = 75% 정확하게 예측함.
```

```python
# 예측의 정확도를 구한다.  예) 100점 만점 중 몇 점을 맞았는지?

(len(y_test) - diff_count) / len(y_test) * 100 # 74& 정확하게 예측
```

```python
# 위에서 처럼 직접 구할수도 있지만 미리 구현된 알고리즘을 가져와 사용한다.

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict) * 100
```

```python
# model의 score로 점수를 계산
model.score(X_test, y_test) * 100
```
